```python
# Load a Text File in Python

def load_file(filename):
    
    with open(filename,encoding='utf-8') as f:
        
        content = f.read().lower()
        return content
```

```python
data = load_file('data.txt')
print(data)
```

```python
# Start Define - Markov Chain - transition table
k = 4

def generateTable(text):
    
    T = {}
    for i in range(len(text)-k):
        X = text[i:i+k]
        Y = text[i+k]
        
      
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
        else:
            if T[X].get(Y) is None:
                T[X][Y] = 1
            else:
                T[X][Y] += 1
                
        #Find the Sum
        
    for kx in T.keys():
        s = 0
        for v in T[kx].values():
            s += v
        #Divde every term in that row by that sum
        for ky in T[kx]:
            T[kx][ky] /= s
                
    return T
```

```python
T =generateTable(data)
```

```python
print(T)
```

```python
def predict(ctx):
    #ctx is the window of last k chars
    
    if T.get(ctx) is None:
        return " "
    
    possible_keys = list(T[ctx].keys())
    possible_values = list(T[ctx].values())
    
    return np.random.choice(possible_keys,p=possible_values)
    
    
    
```
