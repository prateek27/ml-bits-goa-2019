```python
def dist(p1,p2):
    return np.sum((p1-p2)**2)**.5

def knn(X,Y,test_point,k=5):
    
    #Total Points
    m = X.shape[0]
    l = []
    #Dist of Test POint from each point in training data
    for i in range(m):
        dis = dist(test_point,X[i])
        l.append((dis,Y[i]))
        
    l.sort()
    l = np.array(l[:k])
    l = l[:,1]
    #print(l.shape)
    #print(l)
    classes,cnts = np.unique(l,return_counts=True)
    print(classes)
    idx = np.argmax(cnts)
    #class having majority
    pred = classes[idx]
    
    return int(pred)
    
    
    
```

```python
knn(X,Y,test_point)
```
